```python
import pandas as pd
```

```python
df = pd.read_csv("행정안전부_승강기중대사고내역_20240731.csv")
```

```python
# '사고구분' 열 삭제
df = df.drop(columns=['사고구분'])
```

```python
# 결측값이 있는 행 제거
df = df.dropna()
```

```python
# # 각 열의 고유값 확인
# for column in df.columns:
#     print(f"{column}의 고유값:")
#     print(df[column].unique())
#     print("\n")

```

```python
# '발생일자' 열을 datetime 형식으로 변환하고, 2024년만 필터링
df['발생일자'] = pd.to_datetime(df['발생일자'])
df2 = df[df['발생일자'].dt.year == 2023]
```

```python
# '주소' 열에서 '서울특별시'가 포함된 행만 선택

# df3 = df2[df2['주소'].str.contains('서울특별시')]
df3 = df[df['주소'].str.contains('서울특별시')]
```

```python

# 데이터프레임 전체를 출력할 수 있도록 설정 변경
pd.set_option('display.max_rows', None)  # 모든 행 출력
pd.set_option('display.max_columns', None)  # 모든 열 출력
pd.set_option('display.width', None)  # 줄 너비 제한 없음
pd.set_option('display.max_colwidth', None)  # 열 너비 제한 없음

df3
```

```python
# 주소 열에서 괄호가 있는 부분부터 삭제하고, '지하'가 있는 부분도 삭제
df3['주소'] = df3['주소'].str.split('(').str[0].str.strip()  # 괄호 제거
df3['주소'] = df3['주소'].str.split('지하').str[0].str.strip()  # '지하' 제거
```

```python
# # 인덱스 0의 건물명 수정
# df3.at[75, '주소'] = '수정된 건물명'
```

```python
# 인덱스를 번호순으로 재설정하고 기존 인덱스는 삭제
df3.reset_index(drop=True, inplace=True)
df3.index = df3.index + 1 #1번부터 표시되도록
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

# Nominatim 지오코더 초기화
geolocator = Nominatim(user_agent="geo_converter")

# 원본 데이터프레임 df3의 복사본 생성
df4 = df3.copy()

# 위도와 경도를 저장할 열 추가
df4['위도'] = None
df4['경도'] = None

# 각 주소를 반복하면서 위도와 경도를 가져옴
for index, row in df4.iterrows():
    address = row['주소']
    location = geolocator.geocode(address)
    if location:
        df4.at[index, '위도'] = location.latitude
        df4.at[index, '경도'] = location.longitude

# 결과 출력
print(df4)
```

```python
# 결측값이 있는 행 제거
df4 = df4.dropna()
```

```python
import folium

# 첫 번째 주소의 위도와 경도로 지도의 중심을 설정
if not df4.empty:
    center_lat = df4['위도'].mean()
    center_lon = df4['경도'].mean()
else:
    center_lat, center_lon = 0, 0  # 데이터프레임이 비어있을 경우 기본값 설정

# 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=12)

# 각 주소에 빨간색 점 추가
for index, row in df4.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=1,  # 점의 크기
        color='red',  # 점의 색상
        fill=True,
        fill_color='red',
        fill_opacity=0.6,
        popup=row['주소']  # 클릭 시 주소 표시
    ).add_to(m)

# 지도 출력
m.save("df3_location_map.html")
m
```
